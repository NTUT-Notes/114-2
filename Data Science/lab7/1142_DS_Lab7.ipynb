## Simple Linear Regression

We will start with the most familiar linear regression, a straight-line fit to data.
A straight-line fit is a model of the form
$$
y = a + bx
$$
where $b$ is commonly known as the *slope*, and $a$ is commonly known as the *intercept*.

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns; sns.set()
```

```python
x = pd.DataFrame([55,60,65,75,70],columns=['Weight(KG)'])
y = pd.DataFrame([150,170,170,180,180],columns=['Size(CM)'])
plt.scatter(x, y);
```

```python
from sklearn import linear_model
reg=linear_model.LinearRegression()
reg.fit(x,y)
reg.score(x,y)
#predictions=reg.predict(80)
print(reg.coef_)
print(reg.intercept_)
print(reg.score(x,y))
```

```python
x_fit = np.linspace(50, 90, 2)  # 繪圖用的x軸座標 (重量KG)
y_pred = reg.predict(x_fit[:, np.newaxis])  # 使用 x軸座標值 預測 y (高度CM) 值
plt.scatter(x, y)
plt.plot(x_fit, y_pred, 'r');    # 用紅線標示
```

```python
rng = np.random.RandomState(1)
x = 10 * rng.rand(50)
y = 2 * x - 5 + rng.randn(50)
plt.scatter(x, y);
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression(fit_intercept=True) # 一般採用預設參數設置，可根據自己實際要求設置參數
model.fit(x[:, np.newaxis], y) # 訓練

x_fit = np.linspace(0, 10, 1000)  # 繪圖用的x軸座標
y_pred = model.predict(x_fit[:, np.newaxis])  # 使用 x軸座標值 預測 y 值

plt.scatter(x, y)
plt.plot(x_fit, y_pred, 'r');    # 用紅線標示
```

```python
print("Model slope:    ", model.coef_)
print("Model intercept:", model.intercept_)
```
