## Problem 
請用 "date = "2026/03/16" 分析股市的開盤價和收盤價的平均值和標準差。

```python
import pandas as pd
import numpy as np
import requests
```

```python
date = "20260316"
url = f'https://www.twse.com.tw/exchangeReport/MI_INDEX?response=json&date={date}&type=ALLBUT0999'
response = requests.get(url, verify=False)  # 忽略 SSL 證書驗證
response_json = response.json()

# 找到每日收盤行情的數據
stock_data = []
stock_fields = []
for table in response_json['tables']:
    if "每日收盤行情" in table['title']:  # 確保是股票成交資訊
        stock_data = table['data']  # 股票成交數據
        stock_fields = table['fields']  # 欄位名稱
        break

# 確保有數據才建立 DataFrame
if stock_data and stock_fields:
    stock = pd.DataFrame(stock_data, columns=stock_fields)
else:
    print("Error: 未找到每日收盤行情的數據")

stock.head()
```

```python
def convert_to_numeric(price):
    valid=(price!='--')

    num_price = pd.to_numeric(
        pd.Series(price).astype(str).str.replace(',', '', regex=False),
        errors='coerce'
    ).to_numpy()

    return num_price[valid].astype(float)
        
```

### 結果顯示

```python
open_dayprice = convert_to_numeric( np.array(stock.開盤價) )
close_dayprice = convert_to_numeric( np.array(stock.收盤價) )

data = {
    '類別': ["開盤", "收盤",],
    '平均值': [open_dayprice.mean(), close_dayprice.mean()],
    '標準差': [ open_dayprice.std(), close_dayprice.std()]
}

pd.DataFrame(data)
```
