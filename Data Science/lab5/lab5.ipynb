題目一：請繪製台積電在2025年的收盤價及開盤價的趨勢圖。

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
#在Jupyter Notebook上顯示圖表所必需的　magic command,
%matplotlib inline
```

```python
data = pd.read_csv('data_science_data2025.csv',index_col=0)
data_TSMC = data[data['證券代號']=='2330']
data_TSMC = data_TSMC[['date', '開盤價', '收盤價']]

data_TSMC['開盤價'] = data_TSMC['開盤價'].replace({r'$':'', ',':'', '--':np.nan}, regex=True)
data_TSMC['開盤價'] = data_TSMC['開盤價'].astype('float64')

data_TSMC['收盤價'] = data_TSMC['收盤價'].replace({r'$':'', ',':'', '--':np.nan}, regex=True)
data_TSMC['收盤價'] = data_TSMC['收盤價'].astype('float64')

data_TSMC = data_TSMC.set_index('date')

data_TSMC = data_TSMC.rename(columns = {'開盤價':'opening_price', '收盤價':'closing_price'})

data_TSMC.plot()
```
