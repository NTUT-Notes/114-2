### Problem 1

Data: 14.7, 14.4, 8.2, 10.7, 14.6, 14.1, 14.4, 14.4, 18.2, 14.5, 14.5, 14.7, 14.9, 15.1, 15.9, 25.0

(a) Draw the box-and-whisker plot of the following data in Jupyter Notebook.

```python
import matplotlib.pyplot as plt
import numpy as np

data = [14.7, 14.4, 8.2, 10.7, 14.6, 14.1, 14.4, 14.4, 18.2, 14.5, 14.5, 14.7, 14.9, 15.1, 15.9, 25.0]

q1 = np.percentile(data, 25)
q3 = np.percentile(data, 75)
iqr = q3 - q1
upper_outliner = q3 + 1.5 * iqr
lower_outliner = q1 - 1.5 * iqr
upper_extreme = q3 + 3 * iqr
lower_extreme = q1 - 3 * iqr

plt.figure(figsize=(20, 3))

plt.text(upper_outliner, 1.1, "Outliner", color='blue')
plt.axvline(x=upper_outliner, color='blue', linestyle='--', linewidth=1.5)
plt.axvline(x=lower_outliner, color='blue', linestyle='--', linewidth=1.5)

plt.text(upper_extreme, 1.2, "Extreme", color='red')
plt.axvline(x=upper_extreme, color='red', linestyle='--', linewidth=1.5)
plt.axvline(x=lower_extreme, color='red', linestyle='--', linewidth=1.5)

for i in data:
    plt.text(i, 0.9, i, color='black')


plt.boxplot(data, vert=False) 

plt.show()
```

Use Jupyter to write a function of calculating "Factorial" for an input integer “??". Add some function usage information.

```python
def Factorial(a):
    """"輸入一個整數 a ，並回傳 a階乘 (a!) 的值"""
    
    if (a < 0):
        return "沒有負壹階"
    
    if (a <= 1):
        return 1
    
    return a * Factorial(a-1)
```
