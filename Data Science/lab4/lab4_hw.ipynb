請問2026/3/21日台灣股市收盤價每一股超過1000元的股票有幾家公司? 

這幾家每股超過1000元的平均收盤價是多少? 

目前股王是哪一家公司?

```python
import pandas as pd
import numpy as np
import requests
```

```python
date = "20260323"
url = f'https://www.twse.com.tw/exchangeReport/MI_INDEX?response=json&date={date}&type=ALLBUT0999'
response = requests.get(url)
response_json = response.json()
# 找到每日收盤行情的數據
stock_data = []
stock_fields = []
for table in response_json['tables']:
    if "每日收盤行情" in table['title']:  # 確保是股票成交資訊
        stock_data = table['data']  # 股票成交數據
        stock_fields = table['fields']  # 欄位名稱
        break

# 確保有數據才建立 DataFrame
if stock_data and stock_fields:
    stock = pd.DataFrame(stock_data, columns=stock_fields)
else:
    print("Error: 未找到每日收盤行情的數據")
```

```python
replace_map = {",":"", "--":np.nan}
stock = stock.replace(replace_map, regex=True)

stock.收盤價 = pd.to_numeric(stock.收盤價, errors='coerce', downcast='float')

over_1000 = stock.loc[stock.收盤價 > 1000]

print(f"收盤價超過 1000 元共有 {over_1000.size}")
print(f"收盤價超過 1000 元的公司平均盤價為 {np.average(over_1000.收盤價)}")
la_king = over_1000.sort_values(by="收盤價", axis=0).head()

print("股王：")
print(f"{la_king.iloc[0]}")
```
