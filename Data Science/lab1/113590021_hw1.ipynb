Problem 3:  Use Jupyter to write a function of calculating "Factorial" for an input integer "a". Add some function usage information. 

```python
def Factorial(a):
    """"輸入一個整數 a ，並回傳 a階乘 (a!) 的值"""
    
    if (a < 0):
        return "沒有負壹階"
    
    if (a <= 1):
        return 1
    
    return a * Factorial(a-1)
```
